# Security Setup - Users, Groups & Permissions

This notebook creates 2 groups (**Data_Engineers** and **Data_Analysts**), adds 2 users (one per group), and grants appropriate Unity Catalog permissions on the `formula1` catalog.

**Project Structure:**
- Catalog: `formula1`
- Schemas: `landing`, `bronze`, `silver`, `gold`

**Access Model:**
| Group | Access Level | Schemas |
|-------|-------------|----------|
| Data_Engineers | Full modify (SELECT, MODIFY, CREATE TABLE, CREATE VOLUME) | All schemas |
| Data_Analysts | Read-only (SELECT) | silver, gold |

## Create Groups



```python
groups = [g['name'] for g in spark.sql("SHOW GROUPS").collect()]

if "Data_Engineers" not in groups:
    spark.sql("CREATE GROUP `Data_Engineers`")
```

```python
groups = [g['name'] for g in spark.sql("SHOW GROUPS").collect()]

if "Data_Analysts" not in groups:
    spark.sql("CREATE GROUP `Data_Analysts`")
```

## Grant Permissions - Data Engineers

Data Engineers get **full modify access** on all schemas:
- `USE CATALOG` - Access the catalog
- `USE SCHEMA` - Access all schemas
- `SELECT` - Read data from all tables
- `MODIFY` - Insert, update, delete data
- `CREATE TABLE` - Create new tables
- `CREATE VOLUME` - Create volumes (landing schema)

```python
%sql
-- USE CATALOG
GRANT USE CATALOG ON CATALOG formula1 TO `Data_Engineers`;

-- USE SCHEMA on all schemas
GRANT USE SCHEMA ON SCHEMA formula1.landing TO `Data_Engineers`;
GRANT USE SCHEMA ON SCHEMA formula1.bronze TO `Data_Engineers`;
GRANT USE SCHEMA ON SCHEMA formula1.silver TO `Data_Engineers`;
GRANT USE SCHEMA ON SCHEMA formula1.gold TO `Data_Engineers`;

-- SELECT and MODIFY on all schemas
GRANT SELECT, MODIFY ON SCHEMA formula1.landing TO `Data_Engineers`;
GRANT SELECT, MODIFY ON SCHEMA formula1.bronze TO `Data_Engineers`;
GRANT SELECT, MODIFY ON SCHEMA formula1.silver TO `Data_Engineers`;
GRANT SELECT, MODIFY ON SCHEMA formula1.gold TO `Data_Engineers`;

-- CREATE TABLE and CREATE VOLUME
GRANT CREATE TABLE ON SCHEMA formula1.landing TO `Data_Engineers`;
GRANT CREATE TABLE ON SCHEMA formula1.bronze TO `Data_Engineers`;
GRANT CREATE TABLE ON SCHEMA formula1.silver TO `Data_Engineers`;
GRANT CREATE TABLE ON SCHEMA formula1.gold TO `Data_Engineers`;
GRANT CREATE VOLUME ON SCHEMA formula1.landing TO `Data_Engineers`;
```

## Grant Permissions - Data Analysts

Data Analysts get **read-only access** on production-ready schemas only:
- `USE CATALOG` - Access the catalog
- `USE SCHEMA` - Access silver and gold schemas only
- `SELECT` - Read data (no modification allowed)

> **Note:** Analysts do NOT have access to `landing` or `bronze` schemas, as those contain raw/unprocessed data.

```python
%sql
-- USE CATALOG
GRANT USE CATALOG ON CATALOG formula1 TO `Data_Analysts`;

-- USE SCHEMA (only silver and gold - production ready data)
GRANT USE SCHEMA ON SCHEMA formula1.silver TO `Data_Analysts`;
GRANT USE SCHEMA ON SCHEMA formula1.gold TO `Data_Analysts`;

-- SELECT only (read-only access)
GRANT SELECT ON SCHEMA formula1.silver TO `Data_Analysts`;
GRANT SELECT ON SCHEMA formula1.gold TO `Data_Analysts`;
```

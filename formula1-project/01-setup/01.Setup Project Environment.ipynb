# Set-up the project environment for Formula1 Project

- 1.- Create External Location eberdatabrickscoursedl-formula1 
- 2.- Create Catalog formula1
- 3.- Create Schemas landing, bronze, silver and gold
- 4.- Create Volume Files in the landing schema

```python
%python
dbutils.widgets.removeAll()
```

```python
%sql
create widget text storageName default "eberdatabrickscoursedl";
```

```python
%python
storageName = dbutils.widgets.get("storageName")
```

Access Cloud Storage

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS eberdatabrickscoursedl_formula1
URL 'abfss://formula1@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-course-sc`)  
COMMENT 'External location for the formula1 container';
```

## Create Catalog formula 1

```python
%sql
SHOW CATALOGS;
```

```python
%sql
CREATE CATALOG IF NOT EXISTS formula1
MANAGED LOCATION 'abfss://formula1@${storageName}.dfs.core.windows.net'
COMMENT 'Este es el catalogo principal para el proyecto de Formula1';
```

## Create Schemas

```python
%sql
CREATE SCHEMA IF NOT EXISTS formula1.landing;
CREATE SCHEMA IF NOT EXISTS formula1.bronze
    MANAGED LOCATION 'abfss://formula1@${storageName}.dfs.core.windows.net/bronze';
CREATE SCHEMA IF NOT EXISTS formula1.silver
    MANAGED LOCATION 'abfss://formula1@${storageName}.dfs.core.windows.net/silver';
CREATE SCHEMA IF NOT EXISTS formula1.gold
    MANAGED LOCATION 'abfss://formula1@${storageName}.dfs.core.windows.net/gold';
```

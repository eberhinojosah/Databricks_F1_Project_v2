```python
# Define source_file and table_name
source_file = f"{landing_folder_path}/constructors.json"
table_name = f"{catalog_name}.{bronze_schema}.constructors"
```

### Step 1 - Read the json file using the dataframe reader API

```python
# Define the Schema USING DDL
constructors_shema = """constructorId STRING,
                        name STRING,
                        nationality STRING,
                        url STRING
                        """
```

```python
# Read data from the constructor file
constructors_df = (
    spark.read
        .format('json')
        .schema(constructors_shema)
        .option('mode', 'FAILFAST')
        .load(source_file)
    )
```

### Step 2 - Add Metadata Columns
- Source File
- Ingestion Timestamp


```python
constructors_final_df = add_ingestion_metadata(constructors_df)
```

### Step 3 - Write to bronze delta table

```python
(
    constructors_final_df
    .write
    .format('delta')
    .mode('overwrite')
    .saveAsTable(table_name)
)
```
